# Top Scorers 

```python
import requests 
from bs4 import BeautifulSoup
```

```python
request = requests.get("https://www.premierleague.com/stats/top/players/goals")
request
```

```python
content = request.content
```

```python
soup = BeautifulSoup(content, "html.parser")
```

```python
players = soup.find_all("tr", class_ = "table__row")
players 
```

```python
PlayerName = players[0].find_all("a", class_ = "playerName")[0].text.replace("\n", "").strip()
PlayerName
```

```python
Club = players[0].find_all("td", class_ = "hide-s")[0].text.replace("\n", "").strip()
Club 
```

```python
Nationality = players[0].find_all("span", class_ = "stats__player-country")[0].text.replace("\n", "").strip()
Nationality
```

```python
Goals = players[0].find_all("td", class_ = "stats-table__main-stat")[0].text.replace("\n", "").strip()
Goals
```

```python
import pandas as pd 
```

```python
top_players = pd.DataFrame(columns = ['PlayerName', 'Club', 'Nationality', 'Goals'])
top_players 
```

```python
for player in players: 
    PlayerName = player.find_all("a", class_ = "playerName")[0].text.replace("\n", "").strip()
    Club = player.find_all("td", class_ = "hide-s")[0].text.replace("\n", "").strip()
    Nationality = player.find_all("span", class_ = "stats__player-country")[0].text.replace("\n", "").strip()
    Goals = player.find_all("td", class_ = "stats-table__main-stat")[0].text.replace("\n", "").strip()
    top_players.loc[len(top_players.index)] = [PlayerName, Club, Nationality, Goals]
```

```python
top_players
```

```python
top_players.to_csv("Top Scorers.csv")
```
